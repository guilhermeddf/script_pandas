```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('all_projects_normalized.csv')
```

```python
df.head()
```

```python
def isCI(version):
    if version <= 48:
        return 'NO_CI'
    elif version <= 96:
        return 'EARLY_CI'
    else:
        return 'LATE_CI'
    
df['type'] = df['version'].apply(isCI)
```

```python
df.head()
```

```python
df2 = df
df2['version'] = df2['version'].apply(lambda x : x%48)

d = {0:48}
df2.replace({'version': d}, inplace=True)
df2.version.unique()
```

```python
table = pd.pivot_table(df2, index=['project_name', 'type'], columns=['version'], values='normalized_value')
table.head()
table.to_csv('vdcxx.csv', index=True)
```
